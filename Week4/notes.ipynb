## Logistic regression

This model uses logistic function in order to provide us with output which is the estimated probability of occuring that event (it is in the range of 0 to 1).

Logistic function (Sigmoid function): $ f(x) = \frac{1}{1+e^{-x}} $ 

**The graph of Sigmoid function:**

```python
import math 
def f(x):
    s=[]
    for i in x:
        s.append(1/(1+math.exp(-i)))
    return s

import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-10,10,0.2)
sol=f(x)
plt.figure(figsize=(5,5),dpi=80)
plt.plot(x,sol)
plt.legend(['sigmoid function'])
```
